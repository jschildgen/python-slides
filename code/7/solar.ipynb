```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 25)
```

```python
f_peter = lambda x: 800 * 1.04**x
y_peter = f_peter(x)
```

```python
f_ute = lambda x: 100 * x
y_ute = f_ute(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y_peter, label='Peter')
ax.plot(x, y_ute, label='Ute')
```

```python
import scipy

nullstellen = scipy.optimize.fsolve(lambda x: f_peter(x) - f_ute(x), 0)

nullstellen
```
