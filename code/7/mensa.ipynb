```python
import pandas as pd

data = pd.read_csv("mensa.csv", sep=";", encoding="iso-8859-1", decimal=",")

data.head()
```

```python
# anzahl der zeilen
len(data)
```

```python
data.loc[(data["tag"] == "Fr") & (data["warengruppe"] == "Suppe")]
```

```python
# alles wo warengruppe mit HG anfängt
data.loc[data["warengruppe"].str.startswith("HG")]
```

```python
data["warengruppe"] = data["warengruppe"].str.replace(r'\d+', '', regex=True)

data
```

```python
data["name"] = data["name"].str.replace(r'\([^\)]*\)', '', regex=True)
data
```

```python
data.drop("preis", axis=1, inplace=True)
data
```

```python
(data["bed"] / data["stud"]).mean()
```

```python

data.loc[data["warengruppe"] == "HG", "stud"].mean()
```

```python
# Datum umwandeln
data["datum"] = pd.to_datetime(data["datum"], format="%d.%m.%Y")

data
```

```python
from sklearn import linear_model

model1 = linear_model.LinearRegression()
model2 = linear_model.LinearRegression()

model1.fit(data[["stud"]], data["bed"])
model2.fit(data[["stud", "gast"]], data["bed"])
print(model1.predict(pd.DataFrame({'stud': [6.50]})))
```

```python
data2 = pd.read_csv("mensa2.csv", sep=";", encoding="iso-8859-1", decimal=",")

data2["bed_predict1"] = model1.predict(data2[["stud"]])
data2["bed_predict2"] = model2.predict(data2[["stud", "gast"]])

data2.head()
```

```python
# mean squared error between the predicted and the actual value
from sklearn.metrics import mean_squared_error

mean_squared_error(data2["bed"], data2["bed_predict"])
```

```python
model = linear_model.LinearRegression()

data["len"] = data["name"].str.len()

model.fit(data[["len"]], data["stud"])
print(model.predict(pd.DataFrame({'len': [len("Spaghetti Bolognese mit Rindfleisch")]})))
```
